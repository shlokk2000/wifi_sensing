```python
# import torch
# import torch.nn as nn
# import torch.optim as optim
# import torch.nn.functional as F
import numpy as np
import torchvision
from torchvision import *
from torch.utils.data import Dataset, DataLoader 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import copy
import os, math
import tensorflow as tf
from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix
import itertools
from skimage.transform import resize
from librosa.display import specshow
import re
#import cv2
from os import listdir
from os.path import isfile, join
from sklearn.preprocessing import MinMaxScaler
import re
from skimage.transform import resize
from tensorflow.keras import applications
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D,GlobalAveragePooling2D
from tensorflow.keras.callbacks import TensorBoard,ReduceLROnPlateau,ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from skimage.io import imread
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
#%matplotlib inline
from tensorflow import keras

import tensorflow.keras
from tensorflow.keras import optimizers
#from keras.utils import to_categorical
from tensorflow.keras.layers import Input, Dense, Flatten, Dropout, Conv2D, MaxPooling2D, Activation, concatenate
from keras.layers.advanced_activations import LeakyReLU 
from tensorflow.keras.models import Model

from tensorflow.keras.layers import BatchNormalization


################################################################################
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

###############################################################################
def plot_confusion_matrix_new(predicted_labels_list, y_test_list):
    cnf_matrix = confusion_matrix(y_test_list, predicted_labels_list)
    np.set_printoptions(precision=2)

    # Plot non-normalized confusion matrix
    plt.figure()
    generate_confusion_matrix(cnf_matrix, classes=class_names, title='Confusion matrix, without normalization')
    
    plt.savefig(save_dir +'Confusion_matrix_CV_data'+'.png')
    plt.show()
    
    # Plot normalized confusion matrix
    plt.figure()
    generate_confusion_matrix(cnf_matrix, classes=class_names, normalize=True, title='Normalized confusion matrix')
      
    plt.savefig(save_dir +'Normalised Confusion_matrix_CV_data'+'.png')
    plt.show()  
    
def generate_confusion_matrix(cnf_matrix, classes, normalize=False, title='Confusion matrix'):
    if normalize:
        cnf_matrix = cnf_matrix.astype('float') / cnf_matrix.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.imshow(cnf_matrix, interpolation='nearest', cmap=plt.get_cmap('Blues'))
    plt.title(title)
    plt.colorbar()

    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cnf_matrix.max() / 2.

    for i, j in itertools.product(range(cnf_matrix.shape[0]), range(cnf_matrix.shape[1])):
        plt.text(j, i, format(cnf_matrix[i, j], fmt), horizontalalignment="center",
                 color="white" if cnf_matrix[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
    return cnf_matrix
###############################################################################################
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm / cm.sum(axis=1)[:, np.newaxis]  # cm.astype('float')
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.rcParams.update({'font.size': 14})
    

def find_personid_from_path(filenames):
    return [[filename.split('_')[3]] for filename in filenames]

 
def find_roomid_from_path(filenames):
    return [[filename.split('_')[5]] for filename in filenames]

def find_expnum_from_path(filenames):
    return [int(filename.split('_')[1]) for filename in filenames]

def remove_str_nuc1(filenames):
    rstr = r"_NUC1"
    return [[  re.sub(rstr, "", filename) ] for filename in filenames]

def remove_str_nuc2(filenames):
    rstr = r"_NUC2"
    return [[  re.sub(rstr, "", filename) ] for filename in filenames]


def import_spectrograms(filepaths):
    """
    import all spectrogram in the filepaths as one array
    """
    mats = []
    for p in filepaths:
        print("p:",p)
        arr = csv_to_arr(p)
        arr = normalize_arr(arr,axis=0)        
        arr = resize(arr, (224,224), order=1, preserve_range=True)      
        plt.figure()
        plt.show()
        specshow(arr  ,cmap="jet")
        ##################################################################         
        arr = arr.reshape(-1,*arr.shape)
        mats.append(arr)
    return np.concatenate(mats,axis=0)

def csv_to_arr(path):
    """Import single csv file as numpy.ndarray"""
    arr = pd.read_csv(path, header=None).to_numpy()
    return arr

def normalize_arr(arr,axis=0):
    if axis == 0:
        arr = MinMaxScaler().fit_transform(arr)
    elif axis == 1:
        arr = MinMaxScaler().fit_transform(arr.transpose()).transpose()
    return arr


######################################################################################  
def import_pair_data(directory1,directory2,NUC=['NUC1','NUC2'], modal=['nuc1','nuc2'], return_id=True): 

    """
    import all spectrogram (in pair) in the directory
    """

    print("Importing Data ",end='')
    
    data = {'X1':[],'X2':[],'y':[],'personid':[] ,'roomid':[],'expnum':[]  }  ########  ,'expnum':[]         
    
    label_ls = os.listdir(directory1)   
    print("label_ls:", label_ls)
                  
    
    for label in label_ls:
        print('>',end='')
        # selcting available pairs
        pfiles_amp    = [f.split('.')[0] for f in os.listdir(directory1+'/'+label+'/'+modal[0])] ####22/08###
        pfiles_phdiff = [f.split('.')[0] for f in os.listdir(directory2+'/'+label+'/'+modal[1])] ####22/08###
#         print("pfiles_amp", pfiles_amp)
#         print("pfiles_phdiff", pfiles_phdiff) 
 
        pfiles_amp_new    = remove_str_nuc1(pfiles_amp)     
        pfiles_phdiff_new = remove_str_nuc2(pfiles_phdiff)   
#         print("pfiles_amp_new", pfiles_amp_new)
#         print("pfiles_phdiff_new", pfiles_phdiff_new)
        
        available_pairs = np.intersect1d(pfiles_amp_new,pfiles_phdiff_new).tolist()
#         print("available_pairs", available_pairs)
        
        files_amp    = [directory1+'/'+label+'/'+modal[0]+'/'+pfilename+ '_'+ NUC[0] + '.csv' for pfilename in available_pairs] ####22/08###
        files_phdiff = [directory2+'/'+label+'/'+modal[1]+'/'+pfilename+ '_'+ NUC[1] + '.csv' for pfilename in available_pairs] ####22/08###
#         print("files_amp", files_amp)
#         print("files_phdiff", files_phdiff)
        
        # importing
        X1 = import_spectrograms(files_amp)
#         print("X1", X1)
     
        X2 = import_spectrograms(files_phdiff)        
#         print("X2", X2)
        
        y = np.full(X1.shape[0], label)
        
        person_id = find_personid_from_path(available_pairs) ######################################        
        room_id   = find_roomid_from_path(available_pairs)   ####################################
        expnum    = find_expnum_from_path(available_pairs)   ####################################
        
        
        assert X1.shape[0] == X2.shape[0] == len(person_id)        
        
        
        data['X1'].append(X1)
        data['X2'].append(X2)
        data['y'].append(y)
        data['personid'].append(person_id)
        data['roomid'].append(room_id)
        data['expnum'].append(expnum)        
        
                
    print(" Complete")
    if return_id == True:
        return np.concatenate(data['X1']), np.concatenate(data['X2']),np.concatenate(data['y']), np.concatenate(data['personid']),np.concatenate(data['roomid']),np.concatenate(data['expnum'])        
    
    else:
        return np.concatenate(data['X1']), np.concatenate(data['X2']), np.concatenate(data['y'])
    


def initial_filtering_activities(datas,condition_idx=2, activities=[]):
    for activity in activities:
        datas = where(*datas,condition=(datas[2] != activity))
    return datas



def split_datasets(X1,X2,y,by='personid',split=0.8,stratify=None,id_ls=None,validation_id=5):
    X1 = X1.reshape(*X1.shape,1).transpose(0,3,1,2)
    X2 = X2.reshape(*X2.shape,1).transpose(0,3,1,2)
    
    if by == 'random':
        X1_train, X1_test, X2_train, X2_test, y_train, y_test = train_test_split(X1,X2,y,train_size=split,stratify=stratify, random_state = 42)
    elif by == 'personid':
        X1_train, X2_train, y_train = where(X1,X2,y,condition=(id_ls != validation_id))
        X1_test,  X2_test,  y_test  = where(X1,X2,y,condition=(id_ls == validation_id))
    elif by == 'roomid':
        X1_train, X2_train, y_train = where(X1,X2,y,condition=(id_ls != validation_id))
        X1_test,  X2_test,  y_test  = where(X1,X2,y,condition=(id_ls == validation_id))
    elif by == 'expid':
        X1_train, X2_train, y_train = where(X1,X2,y,condition=(id_ls != validation_id))
        X1_test,  X2_test,  y_test  = where(X1,X2,y,condition=(id_ls == validation_id))
        
        
    return X1_train, X1_test, X2_train, X2_test, y_train, y_test


def select_train_test_dataset(X1_train, X1_test, X2_train, X2_test, y_train, y_test, joint):
    if joint == 'joint':
        X_train = np.concatenate((X1_train,X2_train))
        y_train = np.concatenate((y_train,y_train))
        X_test = np.concatenate((X1_test,X2_test))
        y_test = np.concatenate((y_test,y_test))
    elif joint == 'first':
        X_train = X1_train
        y_train = y_train
        X_test = X1_test
        y_test = y_test
    elif joint == 'second':
        X_train = X2_train
        y_train = y_train
        X_test = X2_test
        y_test = y_test
    else:
        raise ValueError("Must be in {'joint','first','second'}")
    return X_train, X_test, y_train, y_test

    
def where(*arrays,condition):
    """select item in multi arguments of numpy.ndarrays based on condition, return arguments of numpy.ndarrays"""
    return [arr[np.where(condition)] for arr in arrays]


##############################################################################
if __name__ == '__main__':

    val_method = 'personid'  # 'personid': # 'roomid' , 'expid'
    ACTIVITY_SKIP = []     # activities to skip
    load_dir  = 'wifi_csi_spectrograms/spectrogram_amplitude_csi'  # spectrogram_phase_difference_csi******************************* 
    load_dir2 = 'wifi_csi_spectrograms/spectrogram_amplitude_csi'  # spectrogram_phase_difference_csi*******************************
    naming    = 'multi_input_cnn'
    network   = 'alexnet'                    # 'resnet50' #'alexnet' # custom

    class_names = os.listdir(load_dir)  
    print(class_names)
    
    ##IMPORT PAIRED DATA FROM 2 RECEIVERS
    datas = import_pair_data(load_dir,load_dir2, NUC = ['NUC1','NUC2'], modal=['nuc1','nuc2'], return_id=True)

    # filtering activities 
    X1,X2,y,person_id, room_id, exp_num = initial_filtering_activities(datas,activities=ACTIVITY_SKIP)

    del datas

    room_df = pd.DataFrame(room_id, columns=["room_id"])
    person_df = pd.DataFrame(person_id, columns=["person_id"])

    ### replace strings by numbers for room_id and person_id

    room_df.room_id[room_df.room_id == '1'] = 1
    room_df.room_id[room_df.room_id == '2'] = 2

    person_df.person_id[person_df.person_id == 'One'] = 1
    person_df.person_id[person_df.person_id == 'Two'] = 2
    person_df.person_id[person_df.person_id == 'Three'] = 3
    person_df.person_id[person_df.person_id == 'Four'] = 4
    person_df.person_id[person_df.person_id == 'Five'] = 5
    person_df.person_id[person_df.person_id == 'Six'] = 6
    person_df.person_id[person_df.person_id == 'Seven'] = 7

    ### CHOOSE WHICH ROOM or PERSON or EXP_NO TO USE AS TEST DATA
    if val_method == 'personid':
        validation_id = 5 # 1,2,3,4,5,6,7#### person_id: 'One', 'Two' , 'Three', 'Four' , 'Five', 'Six' , 'Seven'
        id_ls = np.array(person_df['person_id']).astype(str).astype(int)

    elif val_method == 'roomid':
        validation_id = 1 #1,2 # room_id: '1' , '2' testing on second room 
        id_ls = np.array(room_df['room_id']).astype(str).astype(int)

    elif val_method == 'expid':
        validation_id = 7 #1,2 # exp_id: '1' , '2' , '3' corresponding to exp001, exp002, exp003, etc
        id_ls = exp_num

    
    # id for loov validation, else randomly spliting data
    if val_method == 'personid':
        X1_train, X1_test, X2_train, X2_test, y_train_, y_test_ = split_datasets(X1,X2,y,by='personid',id_ls=id_ls,validation_id=validation_id)
        
    elif val_method == 'roomid':
        X1_train, X1_test, X2_train, X2_test, y_train_, y_test_ = split_datasets(X1,X2,y,by='roomid',id_ls=id_ls,validation_id=2)
      
    elif val_method == 'expid':
        X1_train, X1_test, X2_train, X2_test, y_train_, y_test_ = split_datasets(X1,X2,y,by='expid',id_ls=id_ls,validation_id=validation_id)
            
    elif val_method == 'random':
        X1_train, X1_test, X2_train, X2_test, y_train_, y_test_ = split_datasets(X1,X2,y,by='random',split=0.8,stratify=y)
     

    del X1
    del X2
    del y
    
    print('X1 train shape:',X1_train.shape, 'X1 test shape:' ,X1_test.shape )


    X1_train_reshaped = np.transpose(X1_train, (0, 2, 3, 1))
    X2_train_reshaped = np.transpose(X2_train, (0, 2, 3, 1))
    
    X1_test_reshaped = np.transpose(X1_test, (0, 2, 3, 1))
    X2_test_reshaped = np.transpose(X2_test, (0, 2, 3, 1))
    

    # plt.figure()
    # plt.show()
    # specshow(X1_train[200,:,:][0]  ,cmap="jet")
    # plt.figure()
    # plt.show()
    # specshow(X2_train[200,:,:][0]  ,cmap="jet")
    
###############################################################################################
 
def create_convolution_layers(input_img):
  model = Conv2D(32, (3, 3), padding='same', input_shape=input_shape)(input_img)
  model = LeakyReLU(alpha=0.1)(model)
  model = MaxPooling2D((2, 2),padding='same')(model)
  model = Dropout(0.25)(model)
  
  model = Conv2D(64, (3, 3), padding='same')(model)
  model = LeakyReLU(alpha=0.1)(model)
  model = MaxPooling2D(pool_size=(2, 2),padding='same')(model)
  model = Dropout(0.25)(model)
    
  model = Conv2D(128, (3, 3), padding='same')(model)
  model = LeakyReLU(alpha=0.1)(model)
  model = MaxPooling2D(pool_size=(2, 2),padding='same')(model)
  model = Dropout(0.4)(model)
    
  return model


def AlexNet(input_img):
    
    #X_input = Input(input_shape)
    
    X = Conv2D(96,(11,11),strides = 4, input_shape=input_shape)(input_img)
    X = BatchNormalization(axis = 3)(X)
    X = Activation('relu')(X)
    
    X = MaxPooling2D((3,3),strides = 2)(X)
    
    X = Conv2D(256,(5,5),padding = 'same')(X)
    X = BatchNormalization(axis = 3 )(X)
    X = Activation('relu')(X)
    
    X = MaxPooling2D((3,3),strides = 2)(X)
    
    X = Conv2D(384, (3,3) , padding = 'same')(X)
    X = BatchNormalization(axis = 3)(X)
    X = Activation('relu')(X)
    
    X = Conv2D(384, (3,3) , padding = 'same' )(X)
    X = BatchNormalization(axis = 3)(X)
    X = Activation('relu')(X)
    
    X = Conv2D(256, (3,3) , padding = 'same' )(X)
    X = BatchNormalization(axis = 3)(X)
    X = Activation('relu')(X)
    
    X = MaxPooling2D((3,3),strides = 2)(X)
    
    #X = Flatten()(X)
    
    #X = Dense(4096, activation = 'relu', name = "fc0")(X)
    
    #X = Dense(4096, activation = 'relu', name = 'fc1')(X) 
    
    #X = Dense(num_classes,activation='softmax',name = 'fc2')(X) #####################################
    
    #model = Model(inputs = X_input, outputs = X, name='AlexNet')    
    return X   #model

############################################################################

# in_channels — How many channels the input image has

as_gray    = True
in_channel = 3

if as_gray:
  in_channel = 1

img_rows, img_cols = 224, 224
num_classes = len(class_names)  

batch_size = 32
epochs     = 15
input_shape          = (img_rows, img_cols , in_channel)
input_img            = Input(shape = input_shape)
img_height,img_width = 224,224 


#batch_size — defines the number of samples to be propagated through the network. 
#Too low and the less accurate the estimate of the gradient will be, 
#too high and it requires more memory to train the network. 
#It is a trade-off between accuracy and speed.
#epochs — an epoch is a single pass through the entire training set while training the network


# labels - convert class vectors to binary class matrices One Hot Encoding
labels_train = y_train_
labels_test  = y_test_


from sklearn import preprocessing
le = preprocessing.LabelEncoder()

le.fit_transform(labels_train.ravel())

num_classes = le.classes_.size
print(list(le.classes_))
LABELS =list(le.classes_)


tra_y  = le.fit_transform(labels_train.ravel())
test_y = le.fit_transform(labels_test.ravel())
        
    
from keras.utils import np_utils
y_train_hot = np_utils.to_categorical(tra_y, num_classes)
y_test_hot  = np_utils.to_categorical(test_y, num_classes)

print('New y_train shape: ', y_train_hot.shape)

#####################################  MODEL #########################


if network =='custom':    

    amp_input = Input(shape=input_shape)
    amp_model = create_convolution_layers(amp_input)

    phdiff_input = Input(shape=input_shape)
    phdiff_model = create_convolution_layers(phdiff_input)    
    
    conv  = concatenate([amp_model, phdiff_model])
    conv  = Flatten()(conv)
    dense = Dense(512)(conv)
    dense = LeakyReLU(alpha=0.1)(dense)
    dense = Dropout(0.5)(dense)

    output = Dense(num_classes, activation='softmax')(dense)

    model = Model(inputs=[amp_input, phdiff_input], outputs=[output])

    opt = tf.keras.optimizers.Adam()

    model.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['acc'])

    
    
elif network =='alexnet':     
    amp_input    = Input(shape=input_shape)
    amp_model    = AlexNet(amp_input)
    
    phdiff_input = Input(shape=input_shape)
    phdiff_model = AlexNet(phdiff_input)
    
    conv = concatenate([amp_model, phdiff_model])
    X = Flatten()(conv)
    X = Dense(4096, activation = 'relu' )(X)    
    X = Dense(4096, activation = 'relu' )(X)     
    output = Dense(num_classes,activation='softmax' )(X) #####################################
    
    model = Model(inputs = [amp_input, phdiff_input], outputs = output )    

    opt = tf.keras.optimizers.Adam()

    model.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['acc'])


elif network =='resnet50':     
    
    amp_input    = applications.resnet50.ResNet50(weights= None, include_top=False, input_shape= (img_height,img_width,1))#3
    amp_model    = amp_input.output
    
    phdiff_input = applications.resnet50.ResNet50(weights= None, include_top=False, input_shape= (img_height,img_width,1))#3
    phdiff_model = phdiff_input.output
    
    conv = concatenate([amp_model, phdiff_model])
    #X = Flatten()(conv)
    
#We can use global pooling or a flatten layer to connect the dimensions of the previous layers with the new layers. With just a flatten layer and a dense layer with softmax we can perform close the model and start making classification.
    
    for layer in phdiff_input.layers: 
        layer._name = layer.name + str('_1') #renaming layers
        
    for layer in amp_input.layers:
        layer.trainable = False
        
    for layer in phdiff_input.layers:
        layer.trainable = False 
    
    #X = GlobalAveragePooling2D()(conv)
    #X = Dropout(0.7)(X)
    
    X = Flatten()(conv)    
    X = Dense(128, activation='relu')(X)
    X = Dense(128, activation='relu')(X)
    X = Dense(128, activation='relu')(X)

    output = Dense(num_classes, activation= 'softmax')(X)
    
    model = Model(inputs = [amp_input.input,phdiff_input.input ], outputs = output)
    model.summary()

    opt = tf.keras.optimizers.Adam()

    model.compile(loss='categorical_crossentropy',
              optimizer=opt,
              metrics=['acc'])
#####################################################################

best_weights_file="weights.best.hdf5"

checkpoint = ModelCheckpoint(best_weights_file, monitor='val_acc', verbose=1, save_best_only=True, mode='max')

callbacks = [checkpoint]


save_dir = 'SAVE_DIR/'

callbacks_list = [
  keras.callbacks.ModelCheckpoint(
      filepath=save_dir  +'best_model_acc.{epoch:02d}-{val_acc:.4f}.h5',
      monitor='val_acc', save_best_only=True)
  ]


history = model.fit([X1_train_reshaped, X2_train_reshaped], y_train_hot,
          batch_size=batch_size,
          epochs=epochs,
          callbacks=callbacks_list,
          verbose=1,
          validation_data=([X1_test_reshaped, X2_test_reshaped], y_test_hot),
          shuffle=True
          )


plt.figure(figsize=(6, 4))
plt.plot(history.history['acc'], 'r', label='Accuracy of training data')
plt.plot(history.history['val_acc'], 'b', label='Accuracy of validation data')
plt.plot(history.history['loss'], 'r--', label='Loss of training data')
plt.plot(history.history['val_loss'], 'b--', label='Loss of validation data')
plt.title('Model Accuracy and Loss')
plt.ylabel('Accuracy and Loss')
plt.xlabel('Training Epoch')
plt.ylim(0)
plt.legend()
plt.show()
  
########################################################################################

final_loss, final_acc = model.evaluate([X1_test_reshaped, X2_test_reshaped], y_test_hot, verbose=1)
print("Final loss: {0:.6f}, final accuracy: {1:.6f}".format(final_loss, final_acc))  


predicted_classes = model.predict([X1_test_reshaped, X2_test_reshaped])

# Generate generalization metrics
_, acc = model.evaluate([X1_test_reshaped, X2_test_reshaped], y_test_hot, verbose=0)
  
  
y_pred_test2 = model.predict([X1_test_reshaped, X2_test_reshaped])  

# Take the class with the highest probability from the test predictions
max_y_pred_test2 = np.argmax(y_pred_test2, axis=1) ### predicted
max_y_test2      = np.argmax(y_test_hot, axis=1)   ###actual

from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix , classification_report
print(classification_report(max_y_test2, max_y_pred_test2))

# precision tp / (tp + fp)
precision = precision_score(max_y_test2, max_y_pred_test2 , average="macro")
#print('Precision: %f' % precision)
# recall: tp / (tp + fn)
recall = recall_score(max_y_test2, max_y_pred_test2 , average="macro")
#print('Recall: %f' % recall)
# f1: 2 tp / (2 tp + fp + fn)
f1 = f1_score(max_y_test2, max_y_pred_test2 , average="macro")
#print('F1 score: %f' % f1)

print('Metric scores for unseen test data:')
print('Accuracy: %f' % acc)
# precision tp / (tp + fp)
precision = precision_score(max_y_test2, max_y_pred_test2 , average="macro")
print('Precision: %f' % precision)
# recall: tp / (tp + fn)
recall = recall_score(max_y_test2, max_y_pred_test2 , average="macro")
print('Recall: %f' % recall)
# f1: 2 tp / (2 tp + fp + fn)
f1 = f1_score(max_y_test2, max_y_pred_test2 , average="macro")
print('F1 score: %f' % f1)


SCORES2 = pd.DataFrame(np.vstack([precision, recall,f1 ] ).transpose(1, 0)) 
SCORES2.to_csv(save_dir +"Overall_Precision_Recall_F1_" + naming +".csv", header=["Precision", "Recall", "F1" ], index=False)


# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(max_y_test2,max_y_pred_test2)
print(cm)

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cm, classes=class_names, normalize=True,
                      title='Normalized confusion matrix')
plt.savefig(save_dir +'normalized_confusion_matrix_cnn+' + naming+'.png')
plt.show()
```
